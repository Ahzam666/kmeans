```python
with open ('info.txt') as f:
    raw_data = f.read()
raw_data
data = raw_data.split('\n')
data
```

```python
dictt = {}
for i in data:
    y = i.split(' ',1)
    dictt[y[0]] = [i.split(' ')[2:] [x:x+4] for x in range(0,len(i.split(" ")[2:]),4)]
dictt.popitem()
dictt
```

```python
data = []
for k in dictt.keys():
    for v in dictt[k]:
        data.append(v)
data
```

```python
X = np.array(data)
train = []
for i in X:
    print(int(i[2])*int(i[3]))
    train.append([int(i[2])*int(i[3])])
train
```

```python
train = np.array(train)
train
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(train)
```

```python
labels = kmeans.labels_
centeroid = kmeans.cluster_centers_
print(labels,centeroid)
```

```python
cars = train[labels == 1]
bikes = train[labels == 0]
```

```python
def box(image_path,arr,kind,output_path):
    import cv2
    image_path = r'C:\Users\Ehzam\Desktop\machine learning\New folder\positives\00001.bmp'
    image = cv2.imread(image_path)
    
    x,y,width,height = int(arr[0]), int(arr[1]), int(arr[2]), int(arr[3])
    if kind == "bikes":
        color = (0,255,255)
        thickness = 2
    if kind == "cars":
        color = (255,255,0)
        thickness = 2
    cv2.rectangle(image,(x,y),(x+width,y+height),color,thickness)
    cv2.imshow('image with box',image)
    cv2.waitKey(1)
    cv2.destroyAllWindows()
    
    cv2.imwrite(output_path,image)
box(r'C:\Users\Ehzam\Desktop\machine learning\New folder\positives\00271.bmp',np.array(['168', '293', '120', '170']),'bikes',r'C:\Users\Ehzam\Desktop\machine learning\New folder')
```

```python
path = r'C:\Users\Ehzam\Desktop\machine learning\New folder\positives'
output_path = '/path/to/output_image.jpg'
for bike in bikes:
    for k in dictt.keys():
        for d in dictt[k]:
            if np.array_equal(d,bike):
                box(path+k.split('/')[1],d,'bikes',output_path)
                break
```

```python
for car in cars:
    for k in dictt.keys():
        for d in dictt[k]:
            if np.array_equal(d,car):
                box(path+k.split('/')[1],d,'cars')
                break

    
```
